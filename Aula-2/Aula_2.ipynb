## Estatísticas descritivas de zHD

Agora vamos interpretar algumas medidas estatísticas importantes da distribuição de $z$.

### 📌 Média
A média representa o valor central da distribuição.  
Ela indica o redshift típico dos objetos da amostra.

### 📌 Variância e desvio padrão
A variância mede o espalhamento dos dados em torno da média.  
O desvio padrão é a raiz da variância e tem a mesma unidade dos dados, sendo mais interpretável.

Valores altos indicam grande dispersão na distribuição de $z$.

### 📌 Skewness (assimetria)
A skewness mede a assimetria da distribuição:

- Skewness > 0 → cauda à direita (mais valores altos raros)
- Skewness < 0 → cauda à esquerda
- Skewness ≈ 0 → distribuição simétrica

### 📌 Kurtosis (curtose)
A kurtosis mede o peso das caudas da distribuição:

- Kurtosis ≈ 0 → semelhante à distribuição normal
- Kurtosis > 0 → caudas pesadas (mais outliers)
- Kurtosis < 0 → distribuição mais achatada

---

Essas métricas ajudam a entender se os dados seguem uma distribuição aproximadamente gaussiana ou se apresentam assimetrias e outliers relevantes.

```python
mean = df_reduzido["zHD"].mean()
variance = df_reduzido["zHD"].var()
std = df_reduzido["zHD"].std()
skew = df_reduzido["zHD"].skew()
kurt = df_reduzido["zHD"].kurt()

print("Média:", mean)
print("Variância:", variance)
print("Desvio padrão:", std)
print("Skewness:", skew)
print("Kurtosis:", kurt)
```

### 📌 Interpretação

#### Média (0.2212)
O valor médio indica que a amostra é dominada por objetos de baixo redshift, ou seja, supernovas relativamente próximas em termos cosmológicos.

#### Variância e desvio padrão
A variância (0.0621) e o desvio padrão (0.2493) mostram que há uma dispersão significativa nos valores de $z$, indicando que o conjunto cobre uma faixa ampla de distâncias.

#### Skewness (2.22)
A assimetria positiva forte indica uma distribuição com cauda longa à direita. Isso significa que há poucos objetos em altos redshifts, enquanto a maioria está concentrada em baixos valores de $z$.

#### Kurtosis (8.57)
A curtose elevada indica caudas pesadas e presença de outliers. Isso sugere que a distribuição se afasta significativamente de uma gaussiana, refletindo efeitos de seleção observacional e a natureza do levantamento.

---

### 📌 Conclusão

A distribuição de $z$ no Pantheon+ não é gaussiana, apresentando forte assimetria e caudas pesadas. Isso é consistente com dados observacionais cosmológicos reais, onde efeitos de seleção e limitações instrumentais influenciam fortemente a distribuição dos redshifts.

## Pairplot: visualizando relações entre variáveis

O `pairplot` é uma ferramenta muito útil para análise exploratória de dados.

Ele cria automaticamente gráficos de dispersão entre todas as combinações de variáveis do DataFrame, permitindo visualizar possíveis correlações, tendências e agrupamentos.

Além disso, na diagonal principal são exibidas as distribuições individuais de cada variável, geralmente na forma de histogramas.

O `pairplot` ajuda a identificar:

- correlações lineares
- distribuições assimétricas
- outliers
- agrupamentos
- padrões não lineares

Essa visualização é especialmente útil em análises estatísticas e aprendizado de máquina, pois fornece uma visão geral das relações entre as variáveis do conjunto de dados.

```python
df_pairplot= df[ ["zHD",'m_b_corr', "MU_SH0ES", "MU_SH0ES_ERR_DIAG"]].rename(columns={"MU_SH0ES": "mu",'m_b_corr':'m_b',"MU_SH0ES_ERR_DIAG": "mu_err"})
```

```python
sns.pairplot(df_pairplot)
```

## Covariância e correlação

A covariância mede como duas variáveis variam juntas.  
Valores positivos indicam que as variáveis tendem a crescer juntas, enquanto valores negativos indicam comportamento oposto.

Entretanto, a covariância depende da escala dos dados, o que dificulta comparações diretas.

Por isso, utilizamos também a matriz de correlação, que normaliza os valores no intervalo entre -1 e 1:

- Correlação próxima de 1 → forte relação positiva
- Correlação próxima de -1 → forte relação negativa
- Correlação próxima de 0 → pouca ou nenhuma relação linear

Essas métricas ajudam a identificar dependências entre as variáveis do DataFrame.

```python
import seaborn as sns
corr = df_reduzido.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlação")
plt.show()
```

```python
cov = df_reduzido.cov()

sns.heatmap(cov, annot=True, cmap="coolwarm")
plt.title("Matriz de Covariância")
plt.show()
```
