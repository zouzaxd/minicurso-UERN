### Redshifts

- **zHD** — Hubble Diagram Redshift  
- **zHDERR** — incerteza em `zHD`  
- **zCMB** — redshift corrigido para o referencial CMB  
- **zCMBERR** — incerteza em `zCMB`  
- **zHEL** — redshift heliocêntrico  
- **zHELERR** — incerteza em `zHEL`

### Magnitudes e distâncias

- **m_b_corr** — magnitude corrigida pelo método de Tripp (1998)

- **m_b_corr_err_DIAG** — incerteza diagonal da magnitude corrigida  

> **WARNING:** não utilizar essas incertezas para ajuste cosmológico.  
> Deve-se utilizar a matriz de covariância completa.

- **MU_SH0ES** — módulo de distância calibrado pelo SH0ES

- **MU_SH0ES_ERR_DIAG** — incerteza diagonal de `MU_SH0ES`

### Coordenadas

- **RA** — Ascensão reta  
- **DEC** — Declinação  
- **HOST_RA** — RA da galáxia hospedeira  
- **HOST_DEC** — DEC da galáxia hospedeira

### Propriedades SALT2

- **x1** — stretch SALT2  
- **x1ERR** — incerteza em `x1`  
- **c** — parâmetro de cor  
- **cERR** — incerteza em `c`

### Ajuste

- **FITCHI2** — qui-quadrado do ajuste  
- **FITPROB** — probabilidade do ajuste  
- **NDOF** — número de graus de liberdade

## Filtros 

Os filtros são ferramentas importantes na análise de dados, pois permitem selecionar apenas as informações de interesse dentro de um conjunto de dados maior. Com isso, podemos investigar subconjuntos específicos e realizar análises de forma mais organizada.

````python
df_sdds = df[df["IDSURVEY"] == 1][
    ["zHD", "RA", "DEC", "MU_SH0ES", "MU_SH0ES_ERR_DIAG"]
]
````
